```python
import tensorflow as tf
import numpy as np
from PIL import Image

tf.compat.v1.logging.set_verbosity(tf.compat.v1.logging.ERROR)
# Load the model
model = tf.keras.models.load_model('/home/sharan/Documents/Final_Year_Project/models/1')
CLASS_NAMES = ['Blast', 'Blight', 'Brownspot', 'Tungro']

# Load and preprocess the image
img = np.array(Image.open('/home/sharan/Downloads/test/resized_image.jpg'))
img = np.expand_dims(img, axis=0)  # add batch dimension

# Make a prediction
pred = model.predict(img)
print (pred[0])
CLASS_NAMES[np.argmax(pred[0])]
```
